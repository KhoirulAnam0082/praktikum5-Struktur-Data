Bubble Sort

```python
# Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        # Set flag to True to check if any swapping occurs in this pass
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        # If no swapping occurred, the array is already sorted
        if not swapped:
            break
arr = [64, 34, 25, 12, 22, 11, 90]
bubble_sort_arr = arr.copy()
bubble_sort(bubble_sort_arr)
print("Bubble Sort:", bubble_sort_arr)
```

Selection Sort

```python
# Selection Sort
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
arr = [70, 32, 20, 10, 6, 30, 47]
selection_sort_arr = arr.copy()
selection_sort(selection_sort_arr)
print("Selection Sort:", selection_sort_arr)
```

Insertion Sort

```python
# Insertion Sort
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
arr = [80, 3, 20, 15, 7, 11, 90]
insertion_sort_arr = arr.copy()
insertion_sort(insertion_sort_arr)
print("Insertion Sort:", insertion_sort_arr)
```
